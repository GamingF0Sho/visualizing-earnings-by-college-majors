```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Load data to pandas dataframe and print first row
recent_grads = pd.read_csv("recent-grads.csv")
first_row = recent_grads.iloc[0]
print(first_row)
```

```python
# Print first five rows
print(recent_grads.head())
```

```python
# Print last five rows
print(recent_grads.tail())
```

```python
# Print summary statistics for each column of the dataframe
print(recent_grads.describe())
```

```python
# Drop any rows with null values and print length of a column before and after to determine any change.
raw_data_count = len(recent_grads["Rank"])
recent_grads = recent_grads.dropna()
cleaned_data_count = len(recent_grads["Rank"])
print(raw_data_count)
print(cleaned_data_count)
```

```python
# Create Scatter Plot: Sample Size vs. Median Income
recent_grads.plot(x='Sample_size', y='Median', kind='scatter')
plt.show()
```

```python
# Create Scatter Plot: Sample Size vs. Unemployment Rate
recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter')
plt.show()
```

```python
# Create Scatter Plot: Employed Full Time vs. Median Income
recent_grads.plot(x='Full_time', y='Median', kind='scatter')
plt.show()
```

```python
# Create Scatter Plot: Women as share of total vs. Unemployment Rate
recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter')
plt.show()
```

```python
# Create Scatter Plot: Male Graduates vs. Median Income
recent_grads.plot(x='Men', y='Median', kind='scatter')
plt.show()
```

```python
# Scatter Plot: Women Graduates vs. Median Income
recent_grads.plot(x='Women', y='Median', kind='scatter')
plt.show()
```

```python
# Create Histogram Plot: Sample Size
recent_grads['Sample_size'].hist(bins=25, range=(0,5000))
plt.show()
```

```python
# Create Histogram Plots: Sample_size, Median, Employed, Full_time, ShareWomen, Unemployment_rate, Men, Women

cols = ["", "Sample_size", "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate", "Men", "Women"]

fig = plt.figure(figsize=(8,24))
for r in range(1,9):
    ax = fig.add_subplot(4,2,r)
    ax = recent_grads[cols[r]].plot(kind='hist', rot=40, title=cols[r])
plt.show()
```

```python
# Create Scatter Matrix Plot: Sample Size vs. Median Income
from pandas.tools.plotting import scatter_matrix as sm
sm(recent_grads[['Sample_size', 'Median']], figsize=(10,10))
```

```python
# Create Scatter Matrix Plot: Sample Size vs. Median Income vs. Unemployment Rate
sm(recent_grads[['Sample_size', 'Median', 'Unemployment_rate']], figsize=(14,14))
```

```python
# Create Bar Plot: Major vs. Women as share of total (For the 10 highest paying majors)
recent_grads[:10].plot.bar(x='Major', y='ShareWomen', legend=False)

# Create Bar Plot: Major vs. Women as share of total (For the 10 lowest paying majors)
recent_grads[163:].plot.bar(x='Major', y='ShareWomen', legend=False)
```

```python
# Create Bar Plot: Major vs. Unemployment Rate (For the 10 highest paying majors)
recent_grads[:10].plot.bar(x='Major', y='Unemployment_rate', legend=False)

# Create Bar Plot: Major vs. Unemployment Rate (For the 10 lowest paying majors)
recent_grads[163:].plot.bar(x='Major', y='Unemployment_rate', legend=False)
```
